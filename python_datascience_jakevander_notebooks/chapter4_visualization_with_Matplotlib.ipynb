### Two Interfaces for the price of One

there are 2 types of interfaces mostly used

1) matlab-style interface.
2) object-oriented interface.

#### Matlab_style interface

```python
plt.figure #create a plot figure

plt.subplot(2,1,1)#(rows,columns,panel number)
plt.plot(x,np.sin(x))

plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
```

#### object oriented style

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig,ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

#### most of the times people uses 2type i.e Object oriented interface

### Scatter plots

```python
x = np.linspace(0,10,30)
y = np.sin(x)

plt.style.use('seaborn-whitegrid')
plt.plot(x,y,'o',c = 'black')
```

```python
#to check all the styles avialable
seaborn.style.available
```

```python
plt.plot(x, y, '-ok')
```

#### you can add more parameters like *marker =  'p', color = 'gray', markersize = 15, linewidth = 4, markerfacecolor = 'white', markeredgecolor = 'gray', markeredgewidth = 2

```python
plt.plot(x, y, '-p', color='gray',
markersize=15, linewidth=4,
markerfacecolor='blue',
markeredgecolor='red',
markeredgewidth=2)
```

plt has a fucntion call plt.scatter which is most efficient way of scatter plots.

```python
plt.scatter(x,y,marker = 'o',c = 'r')
```
