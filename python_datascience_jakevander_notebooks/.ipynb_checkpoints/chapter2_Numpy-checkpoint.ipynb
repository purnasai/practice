### Splitting of arrays

The opposite of concatenation is splitting, which is implemented by the functions
np.split , np.hsplit , and np.vsplit . For each of these, we can pass a list of indices
giving the split points:

```python
x= [1,2,3,99,99,3,2,1]
x1,x2,x3 = np.split(x,[3,5])#x array and 3,5 are split positions
print(x1,x2,x3)
```

```python
grid = np.arange(16).reshape((4, 4))
grid
```

```python
upper, lower = np.vsplit(grid, [2])
print(upper)
print(lower)
```

```python
left, right = np.hsplit(grid, [2])
print(left)
print(right)
```

### Exponent and algorithms

```python
x = [1, 2, 3]
print("x =", x)
print("e^x =", np.exp(x))
print("2^x =", np.exp2(x))
print("3^x =", np.power(3, x))
```

```python
x = [1, 2, 4, 10]
print("x=", x)
print("ln(x)=", np.log(x))
print("log2(x) =", np.log2(x))
print("log10(x) =", np.log10(x))
```

### broadcasting

Recall that for arrays of the same size, binary operations are performed on an
element-by-element basis:

```python
a = np.array([0,1,2])
b = np.array([5,5,5])
a+b
```
