**Supervised learning**:
Models that can predict labels based on labeled training data.

**Classification**
Models that predict labels as two or more discrete categories.

**Regression**
Models that predict continuous labels.

**Unsupervised learning**
Models that identify structure in unlabeled data.

**Clustering**
Models that detect and identify distinct groups in the data.

**Dimensionality reduction**
Models that detect and identify lower-dimensional structure in higher-
dimensional data

```python
#before we work with iris dataset
#lets see example of supervised learning i.e Linear regression
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 10 * rng.rand(50)
y = 2 * x -1+rng.randn(50)
plt.scatter(x,y)
#now you will get data for linear regression
```

```python
#step-1: Choose a class of model.
from sklearn.linear_model import LinearRegression
```

**Hyperparameter**

• Would we like to fit for the offset (i.e., intercept)?

• Would we like the model to be normalized?

• Would we like to preprocess our features to add model flexibility?

• What degree of regularization would we like to use in our model?

• How many model components would we like to use?



```python
#step-2: choose model hyperparameters
#specify that we would like to fit the intercept using the fit_inter
#cept hyperparameter
model = LinearRegression(fit_intercept = True)
model
```

```python
#Step3:Arrange data into a features matrix and target vector.
X = x[:,np.newaxis]
X.shape
```

```python
#step4:fit the model to your data.
model.fit(X,y)
```

In Scikit-Learn, by convention all
model parameters that were learned during the fit() process have trailing
underscores; for example, in this linear model, we have the following:

```python
model.coef_
```

```python
model.intercept_
```

```python
#step5: Predict labels for unknown data
xfit = np.linspace(-1,11)
Xfit = xfit[:,np.newaxis]
yfit = model.predict(Xfit)
plt.scatter(x,y)
plt.plot(xfit,yfit)
```

### working with IRIS dataset

lets work with the IRIS dataset as Supervised model

```python
#sklean has iris dataset inbuilt in it. so dont need to download externally.
from sklearn.datasets import load_iris
iris=load_iris()
iris
```

```python
#you can also download using Seaborn
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(iris, hue = 'species')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(iris)
#though you  drawn pair plot here, its not clear how the data is 
#spread and what are they, hence we used hue = species in previous code cell
#that clearly says the data distribution and spread and types
```

```python
#lets take the data into variables
X_iris = iris.drop('species', axis = 1)
X_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris,y_iris, random_state = 1)
```

```python
#with the data arranged, we can follow our own recipe to predict the label
#choose the model class
from sklearn.naive_bayes import GaussianNB
#instantiate model
model = GaussianNB()
#fit the model to data
model.fit(Xtrain, ytrain)
#predict on new data
ymodel = model.predict(Xtest)
```

Finally, we can use the accuracy_score utility to see the fraction of predicted labels
that match their true value:

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, ymodel)
```

### unsupervised learning example:Iris dimensinality

- let’s take a look at reducing the dimensionality of the Iris data so as to more easily visualize it.

- Recall that the Iris data is four dimensional: there are four features recorded for each sample.

- The task of dimensionality reduction is to ask whether there is a suitable lower dimensional representation that retains the essential features of the data. Often dimensionality reduction is used as an aid to visualizing data; after all, it is much easier to plot data in two dimensions than in four dimensions or higher!

- principal component analysis which is a fast linear dimensionality reduction technique. We will ask the model to return two components—that is, a two-dimensionalrepresentation of the data.


```python
#as we already have data arranged
from sklearn.decomposition import PCA
model = PCA(n_components = 2)
model.fit(X_iris)
X_2D = model.transform(X_iris)
```

```python
iris['PCA1'] = X_2D[:,0]
iris['PCA2'] = X_2D[:,1]
sns.lmplot('PCA1','PCA2', hue = 'species', data = iris, fit_reg = False)
```
